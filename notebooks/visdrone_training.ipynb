# ISR Pipeline — YOLO11 VisDrone Training
Fine-tunes YOLO11 on the VisDrone dataset for aerial vehicle detection.

```python
# Step 2 — Verify GPU is available
import torch
print('GPU available:', torch.cuda.is_available())
print('Device:', torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'CPU')
```

```python
# Step 3 — Train YOLO11 on VisDrone
# Ultralytics will automatically download the VisDrone dataset on first run
from ultralytics import YOLO

model = YOLO('yolo11s.pt')

model.train(
    data='VisDrone.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,
    project='runs/train',
    name='visdrone_yolo11s',
)
```

```python
# Step 4 — Evaluate the trained model
metrics = model.val()
print('mAP50:', metrics.box.map50)
print('mAP50-95:', metrics.box.map)
```

```python
# Step 5 — Download the trained weights
# After this runs, download best.pt from the files panel on the left
# Path: runs/train/visdrone_yolo11s/weights/best.pt
# Then drop it into the models/ folder in the repo
import os
weights_path = 'runs/train/visdrone_yolo11s/weights/best.pt'
print('Weights saved at:', os.path.abspath(weights_path))
```
